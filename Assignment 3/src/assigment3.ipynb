# Load Boston Dataset

```python
import numpy as np
import matplotlib.pyplot as plt 
import sklearn
from sklearn.ensemble import RandomForestRegressor
from sklearn.decomposition import PCA
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score, mean_squared_error
from sklearn.preprocessing import MinMaxScaler
from sklearn.cross_validation import train_test_split
import pandas as pd  
import seaborn as sns 

%matplotlib inline

from sklearn.datasets import load_boston

boston_dataset = load_boston()
```

## 1. Load data into pandas

```python
boston = pd.DataFrame(boston_dataset.data, columns=boston_dataset.feature_names)
boston.head()
```

### 1a. Add target to dataset

```python
boston['MEDV'] = boston_dataset.target
boston.head()
```

### 1b. Check for null values

```python
boston.isnull().values.any()
```

## 2. Analyze data

### 2a. Plot a histogram

```python
# set the size of the figure
sns.set(rc={'figure.figsize':(11.7,8.27)})

# plot a histogram showing the distribution of the target values
df = pd.DataFrame(boston['MEDV'])
hist = df.hist(bins=100)
```

### 2b. Pairwise correlation

```python
# compute the pair wise correlation for all columns
corr = boston.corr()
corr.head()
```

## 3. Linear Regression

### 3a. Scale the data

```python
scaler = MinMaxScaler()
scaled_df = scaler.fit_transform(boston)
boston_scale = pd.DataFrame(scaled_df, columns=boston.columns)
boston_scale.head()
```

### 3b. Train the model

```python
X_train, X_test, y_train, y_test = train_test_split(boston, boston.MEDV, test_size=0.2)
regr = LinearRegression()
regr.fit(X_train, y_train)
```

### 3c. Evaluate the model

```python
# model evaluation for training set
r2_train = regr.score(X_train, y_train)
rmse_train = np.sqrt(mean_squared_error(y_train, regr.predict(X_train)))

# model evaluation for testing set
r2_test = regr.score(X_test, y_test)
rmse_test = np.sqrt(mean_squared_error(y_test, regr.predict(X_test)))

# root mean square error of the model
print('RMSE Train: ', rmse_train)
print('RMSE Test: ', rmse_test)

# r-squared score of the model
print('R2 Train: ', r2_train)
print('R2 Test: ', r2_test)
```

### 3d. Plot the model

```python
plt.scatter(y_test, regr.predict(X_test))
```

## 4. PCA

### 4a. Initialize PCA

```python
pca = PCA(n_components = 2)

principalComponents = pca.fit_transform(boston)

principalDf = pd.DataFrame(data = principalComponents, columns = ['pc1', 'pc2'])

finalDf = pd.concat([principalDf, pd.DataFrame(data=boston.MEDV, columns=['target'])], axis = 1)
finalDf.head()
```

### 4b. Plot PCA

```python
plt.scatter(finalDf['pc1'], finalDf['pc2'])
```

### 4c. Train and evaluate model

```python
X_train, X_test, y_train, y_test = train_test_split(finalDf, boston_dataset.target, test_size=0.2)

regr = LinearRegression()
regr.fit(X_train, y_train)

# model evaluation for training set
r2_train = regr.score(X_train, y_train)
rmse_train = np.sqrt(mean_squared_error(y_train, regr.predict(X_train)))

# model evaluation for testing set
r2_test = regr.score(X_test, y_test)
rmse_test = np.sqrt(mean_squared_error(y_test, regr.predict(X_test)))

# root mean square error of the model
print('RMSE Train: ', rmse_train)
print('RMSE Test: ', rmse_test)

# r-squared score of the model
print('R2 Train: ', r2_train)
print('R2 Test: ', r2_test)
```

### 4d. Plot model

```python
plt.scatter(y_test, regr.predict(X_test))
```

## 5. Random Forest

```python
X_train, X_test, y_train, y_test = train_test_split(boston, boston_dataset.target, test_size=0.2)

rf = RandomForestRegressor(n_estimators=500, oob_score=True, random_state=0)
rf.fit(X_train, y_train)

# model evaluation for training set
r2_train = rf.score(X_train, y_train)
rmse_train = np.sqrt(mean_squared_error(y_train, rf.predict(X_train)))

# model evaluation for testing set
r2_test = rf.score(X_test, y_test)
rmse_test = np.sqrt(mean_squared_error(y_test, rf.predict(X_test)))

# root mean square error of the model
print('RMSE Train: ', rmse_train)
print('RMSE Test: ', rmse_test)

# r-squared score of the model
print('R2 Train: ', r2_train)
print('R2 Test: ', r2_test)
```
